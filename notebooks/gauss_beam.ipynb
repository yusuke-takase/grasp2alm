```python
%load_ext autoreload
%autoreload 2

import grasp2alm as g2a
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
from grasp2alm import BeamGauss
```

```python
lmax = 500
mmax = lmax
fwhm_deg = 5.0

ellipticity = 0.5
psi_ell_rad = 0.0
psi_pol_rad = 0.0
cross_polar_leakage = 0.0

Gauss = BeamGauss(fwhm_deg)
alm = Gauss.get_alm(lmax, mmax, ellipticity, psi_ell_rad, psi_pol_rad, cross_polar_leakage)

beam_map = Gauss.get_profile(128)
```
